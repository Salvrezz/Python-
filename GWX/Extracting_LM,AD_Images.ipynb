```python
import os
import pandas as pd
import requests

# Load Excel
file_path = "C:\Users\GWX-DATA\Downloads\S-west\5c0cbcde-de81-4d0f-aba2-8e3cf0394e65.csv
df = pd.read_excel(r"C:\Users\GWX-DATA\Downloads\S-west\5c0cbcde-de81-4d0f-aba2-8e3cf0394e65.csv", sheet_name="")

# Create folders for images
os.makedirs("landmark_images", exist_ok=True)
os.makedirs("address_images", exist_ok=True)

# Function to download image
def download_image(url, save_path):
    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            with open(save_path, "wb") as f:
                f.write(response.content)
    except Exception as e:
        print(f"Failed to download {url}: {e}")

# Loop through rows and download
for _, row in df.iterrows():
    sn = row["SN"]
    name = str(row["Name"]).replace(" ", "_")

    # Landmark image
    if pd.notna(row["LandmarkImage"]):
        landmark_path = f"landmark_images/{sn}_{name}.jpg"
        download_image(row["LandmarkImage"], landmark_path)

    # Address image
    if pd.notna(row["AddressImage"]):
        address_path = f"address_images/{sn}_{name}.jpg"
        download_image(row["AddressImage"], address_path)

print("Download complete. Images are saved in 'landmark_images/' and 'address_images/'.")
```
